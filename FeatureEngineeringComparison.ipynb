```
import pickle
import matplotlib.pyplot as plt
```

```
hist_no_feat_path = '/content/drive/My Drive/IncubatorCVProject/models/TwoLayerCNN_no_feat/history.pkl'
hist_w_feat_path = '/content/drive/My Drive/IncubatorCVProject/models/TwoLayerCNN_w_feat/history.pkl'
```

```
with open(hist_no_feat_path,'rb') as file:
    hist_no_feat = pickle.load(file)
with open(hist_w_feat_path,'rb') as file:
    hist_w_feat = pickle.load(file)
    
```

```
hist_no_feat.keys()
```

```
for key in ['train_loss', 'train_acc', 'valid_loss', 'valid_acc']:
    print(key)
    line1, = plt.plot(hist_no_feat[key])
    line2, = plt.plot(hist_w_feat[key])
    plt.legend([line1,line2],['no features','with features'])
    plt.title(key +' vs epoch')
    plt.xlabel('epoch')
    plt.ylabel(key)
    plt.show()
    
```
